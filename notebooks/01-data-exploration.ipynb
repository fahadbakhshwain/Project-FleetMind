```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set some styling for our plots
sns.set_style('darkgrid')

print("Libraries imported successfully!")
```

```python
# Load the dataset
# We use a more robust path starting from the project root
import os

# Define the path to the data file
# os.getcwd() gets the current working directory, we go up one level to the project root
# NOTE: This assumes the notebook is in a subfolder like 'notebooks'
file_path = os.path.join(os.getcwd(), '..', 'data', 'raw', 'hour.csv')

# Let's try a simpler path first, assuming the working directory is the project root
file_path_simple = 'data/raw/hour.csv'

try:
    df = pd.read_csv(file_path_simple)
    print("File loaded successfully using simple path!")
except FileNotFoundError:
    print(f"Could not find file at: {file_path_simple}")
    print("Trying alternative path...")
    try:
        df = pd.read_csv(file_path)
        print("File loaded successfully using alternative path!")
    except FileNotFoundError:
        print("Could not find the file in either path. Please check the file location and name.")
        
# If the dataframe 'df' was created, display the first 5 rows
if 'df' in locals():
    display(df.head())
```

```python
# Get a concise summary of the dataframe
print("DataFrame Info:")
df.info()
```

```python
# Get descriptive statistics for numerical columns
print("\nDescriptive Statistics:")
display(df.describe())
```

```python
# Check for missing values in each column
print("\nMissing Values per Column:")
print(df.isnull().sum())
```

```python
# Convert 'dteday' column to datetime objects
df['dteday'] = pd.to_datetime(df['dteday'])

# Verify the change by checking the dtypes again
print(df.info())
```

```python
# Drop columns that are redundant or not useful for prediction
columns_to_drop = ['instant', 'dteday', 'atemp', 'casual', 'registered']
df_cleaned = df.drop(columns=columns_to_drop)

# Display the first 5 rows of the new cleaned dataframe
print("\nCleaned DataFrame:")
display(df_cleaned.head())

# Display the new shape of the dataframe
print(f"\nShape of the cleaned dataframe: {df_cleaned.shape}")
```

```python
# Set the figure size for better readability
plt.figure(figsize=(14, 6))

# Group by hour and calculate the average number of rentals
sns.barplot(x='hr', y='cnt', data=df_cleaned)

# Add titles and labels
plt.title('Average Bike Rentals by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Number of Rentals (cnt)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Group by weekday and calculate the average number of rentals
sns.barplot(x='weekday', y='cnt', data=df_cleaned)

# Add titles
```

```python
# Create a figure with two subplots side-by-side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

# Plot for Season
# Let's map the season numbers to actual names for better readability
season_map = {1: 'Spring', 2: 'Summer', 3: 'Fall', 4: 'Winter'}
df_cleaned['season_label'] = df_cleaned['season'].map(season_map)
sns.barplot(x='season_label', y='cnt', data=df_cleaned, ax=ax1, order=['Spring', 'Summer', 'Fall', 'Winter'])
ax1.set_title('Average Rentals by Season')
ax1.set_xlabel('Season')
ax1.set_ylabel('Average Rentals')

# Plot for Weather Situation
weather_map = {1: 'Clear', 2: 'Mist/Cloudy', 3: 'Light Snow/Rain', 4: 'Heavy Rain/Snow'}
df_cleaned['weather_label'] = df_cleaned['weathersit'].map(weather_map)
sns.barplot(x='weather_label', y='cnt', data=df_cleaned, ax=ax2)
ax2.set_title('Average Rentals by Weather Situation')
ax2.set_xlabel('Weather Situation')
ax2.set_ylabel('Average Rentals')
# Rotate x-axis labels for better fit
ax2.tick_params(axis='x', rotation=15)

plt.tight_layout() # Adjusts plots to prevent overlap
plt.show()
```

```python
# We should use the numerical dataframe before we added labels
# So we use the original df_cleaned and select only numerical columns
numerical_df = df_cleaned.select_dtypes(include=np.number)

# Calculate the correlation matrix
correlation_matrix = numerical_df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(12, 10))

# Draw the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
# annot=True يكتب قيمة البيانات في كل خلية
# cmap='coolwarm' هو نظام ألوان جيد للارتباطات (الإيجابي دافئ، السلبي بارد)
# fmt=".2f" ينسق الأرقام إلى منزلتين عشريتين

plt.title('Correlation Matrix of Numerical Features')
plt.show()
```
